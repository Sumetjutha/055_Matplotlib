```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1.2,3,2,4.5,4.3]
```

```python
size = [100,200,800,100,300]
plt.scatter(x,y,s=size)
```

```python
size = [100,200,800,100,300]
color = ["lightgrey","lightpink","lightblue","yellow","lightgreen"]
plt.scatter(x,y,s=size,c=color)
```
