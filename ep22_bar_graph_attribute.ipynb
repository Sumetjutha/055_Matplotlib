```python
import matplotlib.pyplot as plt
```

```python
course = ["Python","Java","C#"]
```

```python
score = [80,75,50]
```

```python
c = ["lightgreen","orange","lightblue"]
```

```python
plt.bar(course, score,color=c)
plt.xlabel("Course")
plt.ylabel("Score")
plt.title("Score Midterm")
```

```python
plt.bar(course, score,color=c,width=0.6)
plt.xlabel("Course")
plt.ylabel("Score")
plt.title("Score Midterm")
```

```python
plt.bar(course, score,color=c,width=0.6,edgecolor="black",linewidth=3)
plt.xlabel("Course")
plt.ylabel("Score")
plt.title("Score Midterm")
```

```python
plt.barh(course, score,color=c,edgecolor="black",linewidth=3)
plt.xlabel("Score")
plt.ylabel("Course")
plt.title("Score Midterm")
```
