```python
import matplotlib.pyplot as plt
```

```python
product1 = [10,20,30,40,50]
product2 = [15,30,10,20,60]
month = [1,2,3,4,5]
```

```python
plt.plot(month,product1,color="salmon",marker="s",linestyle='--')
plt.plot(month,product2,marker="o",linestyle='-.')
plt.xlabel("Month")
plt.ylabel("Sales")
```

```python
plt.plot(month,product1,color="salmon",marker="s",linestyle='-')
plt.plot(month,product2,marker="o",linestyle=':')
plt.xlabel("Month")
plt.ylabel("Sales")
```

```python
plt.plot(month,product1,'gs--')
plt.plot(month,product2,marker="o",linestyle='-.')
plt.xlabel("Month")
plt.ylabel("Sales")
```
