```python
import matplotlib.pyplot as plt
```

```python
product1 = [10,20,30,40,50]
product2 = [15,30,10,20,60]
month = [1,2,3,4,5]
```

```python
plt.plot(month,product1,color="salmon",marker="s",linestyle='--')
plt.plot(month,product2,marker="o",linestyle='-.')
plt.xlabel("Month",size=15,color='salmon',backgroundcolor='lightgreen')
plt.ylabel("Sales",size=15,color='gold',backgroundcolor='salmon')
```

```python
plt.plot(month,product1,color="salmon",marker="s",linestyle='--')
plt.plot(month,product2,marker="o",linestyle='-.')
data = {'size':20,'color':"red",'backgroundcolor':'yellow'}

plt.xlabel("Month",data)
plt.ylabel("Sales",data)
```
