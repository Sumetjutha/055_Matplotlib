```python
import matplotlib.pyplot as plt
```

```python
product1 = [10,20,30,40,50]
product2 = [15,30,10,20,60]
month = [1,2,3,4,5]
```

```python
plt.plot(month,product1,'gs--')
plt.plot(month,product2,'r*-.')

data = {'size':12,'color':"red",'backgroundcolor':'yellow'}

plt.xlabel("Month",data)
plt.ylabel("Sales",data)

config = {"color":"blue","size":15}
plt.title("Top Sale 2020",config)

plt.text(2,32,"Point 1",size="xx-large",color="orange")
plt.text(3.8,17,"Point 2",size="small",color="red")
```
