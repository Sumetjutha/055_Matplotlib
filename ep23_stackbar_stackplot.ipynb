```python
import matplotlib.pyplot as plt
```

```python
room = ["A","B","C"]
```

```python
boys = [10,15,5]
```

```python
girls = [20,7,15]
```

```python
plt.bar(room,boys,label="Boys",color="lightblue")
plt.bar(room,girls,bottom=boys,label="Boys",color="lightgreen")
plt.legend()
```

```python
plt.stackplot(room,boys,girls,labels=["Boys","Girls"],colors=["lightblue","lightgreen"])
plt.legend()
```
